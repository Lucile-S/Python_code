##### Class StandartScaler2

```python
from sklearn.preprocessing import StandardScaler


class StandartScaler2:
    def __init__(self):
    ## optionnel; on déclare les variables qu'on va utiliser 
        self.X=None
        self.params_= None

    # Compute the mean and std to be used for later scaling
    def fit(self,X):

        self.mean= X.mean(axis=0)
        self.sd = X.std(axis=0)
        self.params_= {'mean': self.mean , 'sd': self.sd}
    
    # Fit to data, then transform it
    def fit_transform(self,X):
        self.fit(X)
        X_scaled=(X-self.params_['mean'])/self.params_['sd']
        return X_scaled
    
    # Scaling of the data
    def transform(self,X):
        NoneType = type(None)
        if isinstance(self.params_, NoneType):
            print("Vous essayez de transformer sans avoir de fit")
        else:
            X_scaled=(X-self.params_['mean'])/self.params_['sd']
            return X_scaled
```
